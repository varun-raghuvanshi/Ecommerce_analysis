```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("Superstore_USA.xlsx")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["Product Base Margin"].fillna(df["Product Base Margin"].mean())
```

```python
df.isnull().sum()
```

```python
df["Order Priority"].value_counts()
```

```python
df["Order Priority"].unique()
```

```python
df["Order Priority"]=df["Order Priority"].replace('Critical ', 'Critical')
```

```python
sns.countplot(data=df,x="Order Priority")
plt.title("Count of Order Priority")
plt.show()
```

```python
plt.savefig("Count of Order Priority")
```

```python
df["Ship Mode"].value_counts()
```

```python
x=df["Ship Mode"].value_counts().index
x
y=df["Ship Mode"].value_counts().values
y
```

```python
plt.pie(y,labels=x,startangle=60,autopct="%0.2f%%")
plt.legend(loc=1)
plt.title("shipmode")
plt.savefig("shipmode")
plt.show()
```
